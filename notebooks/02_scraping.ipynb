Voy a scrapear información acerca de circuitos

Imports

```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
import time
```

1. Generar URL

```python
circuits_df = pd.read_csv("../data/raw/Used/circuits.csv")

circuits_df["wikipedia_url"] = circuits_df["url"]
```

2. Scrapear cada circuito

```python
scraped_circuits = []

for index, row in circuits_df.head(10).iterrows():
    circuitId = row["circuitId"]
    url = row["wikipedia_url"]

    print(f"Scraping {row["name"]}")

    try:
        headers = {"User-Agent": "UniversityProject/1.0 (Educational Purpose)"}
        response = requests.get(url, headers = headers)

        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            infobox = soup.find("table", {"class": "infobox"})

            if infobox:
                circuit_data = {"circuitId": circuitId}
                rows = infobox.find_all("tr")

            for row_html in rows:
                         header = row_html.find("th")
                         value = row_html.find("td")

                         if header and value:
                              key = header.get_text(strip=True)
                              val = value.get_text(strip=True)

                              if "Capacity" in key or "Capacidad" in key:
                                  circuit_data["capacity"] = val

                              elif "Length" in key or "Longitud" in key:
                                    circuit_data["length"] = val

                              elif "Turns" in key or "Curvas" in key:
                                    circuit_data["turns"] = val
                                    
                              elif "Lap record" in key or "Récord de vuelta" in key:
                                    circuit_data["lap_record"] = val

                              elif "Opened" in key or "Inaugurado" in key:
                                    circuit_data["opened_year"] = val
                                    
                              elif "Surface" in key or "Superficie" in key:
                                    circuit_data["surface"] = val

            scraped_circuits.append(circuit_data)
        
        time.sleep(2)

    except Exception as e:
        print(f"Error scraping {url}: {e}")

                    
            
```

3. Guardar datos

```python
df_scraped = pd.DataFrame(scraped_circuits)

df_scraped.info()
df_scraped.head()
```
